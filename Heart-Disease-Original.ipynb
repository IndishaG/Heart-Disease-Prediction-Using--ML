```python
X=Heart_data.drop(columns='target',axis=1)
Y=Heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

**Splitting your dataset into train and test data**



```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=2)

```

```python
print(X.shape,X_train.shape,X_test.shape)
```

 **Model training**

```python
#Training our LogisticRegression model with training data

model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Model evaluation
#Accuracy score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy of the model for training data : " ,training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy of the model for testing data : " ,testing_data_accuracy)
```

```python
#Building a Predictive System

input_data = (59 ,	1 ,	0 ,	165 ,	176 ,	1 ,	0 ,	90 ,	0 ,	1.0 ,	1 ,	2 ,	1)
input_to_numpyarray= np.asarray(input_data)
input_reshaped= input_to_numpyarray.reshape(1,-1)
prediction = model.predict(input_reshaped)
print(prediction)
if(prediction==0):
  print("The person does not have heart disease")
else:
  print("The person has heart disease" )
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)

# Predict the target variable using the trained model
X_train_prediction = dtc.predict(X_train)

# Calculate the accuracy of the model
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("Accuracy:", training_data_accuracy)
```

```python

# Predict the target variable using the test model
X_test_prediction = dtc.predict(X_test)

# Calculate the accuracy of the model
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("Accuracy:", test_data_accuracy)
```
